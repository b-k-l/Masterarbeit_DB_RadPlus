```python
# Evaluation von ausgewählten Fahrradstraßen vorher-nachher, Geschwindigkeit DB Rad+
# Zur Erstellung des Codes wurde die generative Künstliche Intelligenz (KI) „Claude AI“ des Anbieters Anthropic in Version 3.7 genutzt

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pyarrow.parquet as pq
import gc
import os
import time
from collections import defaultdict
from tqdm.notebook import tqdm
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
from scipy import stats
import pyarrow.compute as pc
import json
import re

# Configuration
PARQUET_FILE = "data/network_all_months_plus_25833_length_with_fahrradstrasse.parquet"
OUTPUT_DIR = "analysis_results/008_Fahrradstrassen_Evaluation"
COLUMN_LENGTH = 'length_m'  # Column for segment length in meters
COLUMN_GEOMETRY = 'geometry_wkt'  # Column for WKT geometry

# Weight selection for analysis 
# Options: 'segment_count', 'segment_length', 'route_count', 'route_length'
WEIGHT_TYPE = 'route_length'  # Choose which weighting method to use

# Before-After periods for each Fahrradstraße
FAHRRADSTRASSE_PERIODS = {
    'Wallstraßen': {
        'before': ['23-04', '23-05', '23-06', '23-07', '23-08'],
        'after': ['24-04', '24-05', '24-06', '24-07', '24-08', '24-09', '24-10', '24-11', '24-12']
    },
    'Handjerystraße': {
        'before': ['23-04', '23-05', '23-06', '23-07', '23-08', '23-09'],
        'after': ['24-07', '24-08', '24-09', '24-10', '24-11', '24-12']
    },
    'Stallschreiberstraße': {
        'before': ['23-04', '23-05', '23-06', '23-07', '23-08', '23-09', '23-10', '23-11'],
        'after': ['24-01', '24-02', '24-03', '24-04', '24-05', '24-06', '24-07', '24-08', '24-09', '24-10', '24-11', '24-12']
    },
    'Hönower Weg': {
        'before': ['23-04', '23-05', '23-06', '23-07', '23-08', '23-09', '23-10'],
        'after': ['24-08', '24-09', '24-10', '24-11', '24-12']
    }
}

# Fahrradstraße columns mapping
FAHRRADSTRASSE_COLUMNS = {
    'Wallstraßen': 'fs_wallstr',
    'Handjerystraße': 'fs_handjerystr',
    'Stallschreiberstraße': 'fs_stallschreiberstr',
    'Hönower Weg': 'fs_hoenowerweg'
}

# Plot configuration - all text elements as variables for easy customization
PLOT_CONFIG = {
    # Text elements
    'TITLE_MAIN': 'Evaluation Einrichtung Fahrradstraße {name}: Geschwindigkeitsanalyse DB Rad+ Daten vorher-nachher',
    'TITLE_SUBTITLE': 'Summierte Messwerte: {route_count_before:,} (Vorher), {route_count_after:,} (Nachher) auf {count} Segmenten',
    'TITLE_PERIOD': 'Vor Umbau: {before_period}, Nach Umbau: {after_period}',
    
    'HIST_TITLE': 'Geschwindigkeitshistogramme',
    'HIST_X_LABEL': 'Geschwindigkeit (km/h)',
    'HIST_Y_LABEL': 'Prozentsatz (%)',
    'HIST_BEFORE_LABEL': 'Vorher: {length:.1f} km, {route_count:,} Messwerte',
    'HIST_AFTER_LABEL': 'Nachher: {length:.1f} km, {route_count:,} Messwerte',
    'HIST_BEFORE_MEAN_LABEL': 'Ø vorher: {mean:.2f} km/h',
    'HIST_AFTER_MEAN_LABEL': 'Ø nachher: {mean:.2f} km/h',
    'HIST_MEAN_DIFF': 'Unterschied Ø: {diff:.2f} km/h ({pct_diff:.1f}%)',
    'HIST_WILCOXON': 'Wilcoxon: {p_value}\nCohens d: {cohens_d:.3f} ({effect})',#\n% Segmente langsamer nachher: {pct_reduced:.1f}%',
    
    'DIFF_TITLE': 'Verteilung der Geschwindigkeitsänderungen nach Segmenten',
    'DIFF_X_LABEL': 'Differenz Geschwindigkeit (km/h, Nachher - Vorher)',
    'DIFF_Y_LABEL': 'Anzahl Segmente',
    'DIFF_NO_CHANGE': 'Keine Veränderung',
    'DIFF_MEAN': 'Änderung Ø: {mean:.2f} km/h',
    'DIFF_MEDIAN': 'Änderung Median: {median:.2f} km/h',
    'DIFF_WEIGHTED': 'Längengewichtete Änderung: {weighted:.2f} km/h',
    
    'SCATTER_TITLE': 'Regression Segmentgeschwindigkeiten (Punktgröße = Segmentlänge)',
    'SCATTER_X_LABEL': 'Ø vorher (km/h)',
    'SCATTER_Y_LABEL': 'Ø nachher (km/h)',
    'SCATTER_EQUAL': 'Gleiche Geschwindigkeit',
    'SCATTER_REGRESSION': 'Regression: y={slope:.2f}x+{intercept:.2f}, R²={r_squared:.2f}',
    
    'BOX_TITLE': 'Boxplots vorher - nachher mit gepaarten Segmentgeschwindigkeiten',
    'BOX_Y_LABEL': 'Geschwindigkeit (km/h)',
    'BOX_BEFORE_LABEL': 'Vorher',
    'BOX_AFTER_LABEL': 'Nachher',
    'BOX_WEIGHTED_MEAN': 'Mittelwert: {mean:.2f} km/h',
    'BOX_SAMPLE_SIZE': 'Effektive n: {effective:.1f}\nSegmente: {actual}',
    
    # Colors
    'BEFORE_COLOR': 'blue',
    'AFTER_COLOR': 'green',
    'BEFORE_COLOR_DARK': 'darkblue',
    'AFTER_COLOR_DARK': 'darkgreen',
    'DIFF_COLOR': 'purple',
    'NO_CHANGE_COLOR': 'red',
    'MEAN_COLOR': 'blue',
    'MEDIAN_COLOR': 'green',
    'WEIGHTED_COLOR': 'orange',
    'SCATTER_COLOR': 'purple',
    'BOX_BEFORE_COLOR': 'blue',
    'BOX_AFTER_COLOR': 'green',
    
    # Layout elements
    'FIGURE_SIZE': (12, 8),
    'DPI': 300,
    'HIST_ALPHA_BEFORE': 0.7,
    'HIST_ALPHA_AFTER': 0.5,
    'HIST_BAR_WIDTH': 0.8,
    'SCATTER_ALPHA': 0.6,
    'BOX_ALPHA': 0.6,
    'TEXT_BOX_STYLE': dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8)
}

# Create output directory
os.makedirs(OUTPUT_DIR, exist_ok=True)

def log(message):
    """Print a timestamped log message"""
    print(f"[{time.strftime('%H:%M:%S')}] {message}")

def force_gc():
    """Force garbage collection and log memory usage"""
    try:
        import psutil
        process = psutil.Process(os.getpid())
        before = process.memory_info().rss / 1024 / 1024
        gc.collect()
        after = process.memory_info().rss / 1024 / 1024
        log(f"Memory: {after:.1f}MB (freed {before-after:.1f}MB)")
    except:
        gc.collect()
        log("Garbage collection performed")

def get_route_count_column(month_code):
    """Get the column name for route count based on month code"""
    return f"{month_code}_route_count"

def parse_histogram(hist_str):
    """Parse histogram string to numpy array"""
    try:
        if isinstance(hist_str, str):
            hist_str = hist_str.strip('[]')
            # Fast NumPy parsing
            try:
                values = np.fromstring(hist_str, sep=',')
                return values
            except:
                # Fallback to manual parsing if NumPy method fails
                values = [float(x) for x in hist_str.split(',')]
                return np.array(values)
        elif isinstance(hist_str, (list, np.ndarray)):
            return np.array(hist_str)
        else:
            return np.zeros(32)  # Return zeros for missing histograms
    except Exception as e:
        log(f"Error parsing histogram: {e}")
        return np.zeros(32)

def calculate_weighted_mean(histogram):
    """Calculate mean from a histogram"""
    speed_bins = np.arange(32)  # 0-31 km/h
    total_count = np.sum(histogram)
    
    if total_count > 0:
        mean_speed = np.sum(speed_bins * histogram) / total_count
        return mean_speed
    return np.nan

def calculate_effective_sample_size(weights):
    """Calculate effective sample size using relative weight approach"""
    if len(weights) == 0 or np.sum(weights) == 0:
        return 0
    
    weights = np.array(weights)
    effective_n = np.sum(weights)**2 / np.sum(weights**2)
    return effective_n

def format_period_text(months):
    """Format a list of month codes into a readable period string"""
    # Convert month codes like '23-04' to a readable format
    if not months:
        return "N/A"
    
    # Extract year and month from first and last entries
    start_month = months[0]
    end_month = months[-1]
    
    # Format: MM/YY
    start_formatted = f"{int(start_month.split('-')[1])}/{start_month.split('-')[0]}"
    end_formatted = f"{int(end_month.split('-')[1])}/{end_month.split('-')[0]}"
    
    return f"{start_formatted}-{end_formatted}"

def get_weight_factor(row, period_months, weight_type):
    """
    Calculate the appropriate weight factor based on selected weight type
    
    Parameters:
    -----------
    row : pd.Series
        The row data containing length and route counts
    period_months : list
        List of month codes for the period
    weight_type : str
        'segment_count', 'segment_length', 'route_count', or 'route_length'
        
    Returns:
    --------
    float
        The weight factor to use
    """
    if weight_type == 'segment_count':
        return 1.0  # Equal weight for each segment
    
    elif weight_type == 'segment_length':
        return float(row[COLUMN_LENGTH])  # Weight by segment length
    
    elif weight_type == 'route_count':
        # Sum route counts for all months in period
        route_count = 0
        for month in period_months:
            count_col = get_route_count_column(month)
            if count_col in row and not pd.isna(row[count_col]):
                route_count += row[count_col]
        return max(1.0, route_count)  # Ensure minimum weight of 1
    
    elif weight_type == 'route_length':
        # Weight by both length and route count
        route_count = 0
        for month in period_months:
            count_col = get_route_count_column(month)
            if count_col in row and not pd.isna(row[count_col]):
                route_count += row[count_col]
        return float(row[COLUMN_LENGTH]) * max(1.0, route_count)
    
    else:
        # Default to segment length if unknown weight type
        log(f"Unknown weight type: {weight_type}, defaulting to segment_length")
        return float(row[COLUMN_LENGTH])

def calculate_cohens_d_paired(differences, weights, effective_n=None):
    """Calculate Cohen's d effect size for paired data with weights"""
    # Get weighted mean of differences
    mean_diff = np.average(differences, weights=weights)
    
    # Weighted variance of differences
    var_diff = np.average((differences - mean_diff)**2, weights=weights)
    
    # Standard deviation of differences
    std_diff = np.sqrt(var_diff)
    
    # Cohen's d for paired data
    d = abs(mean_diff) / std_diff if std_diff > 0 else 0
    
    # If effective sample size is provided, use it to adjust Cohen's d
    # This adjustment is sometimes recommended for large sample sizes
    if effective_n is not None and effective_n > 0:
        # Bias correction for Cohen's d
        # This is one approach to adjust effect size based on sample size
        d_corrected = d * (1 - 3/(4*effective_n - 1))
        # For now, let's just use the standard d since the correction is minimal
        pass
    
    return d, mean_diff

def calculate_metrics(dist, speed_bins):
    """Calculate metrics without bootstrap confidence intervals"""
    result = {}
    
    # Mean speed
    mean = np.sum(speed_bins * dist)
    result['mean'] = float(mean)
    
    # Standard deviation
    variance = np.sum(dist * (speed_bins - mean)**2)
    std = np.sqrt(variance)
    result['std'] = float(std)
    
    # Coefficient of variation
    cv = std / mean if mean > 0 else np.nan
    result['cv'] = float(cv) if not np.isnan(cv) else "NaN"
    
    # Skewness
    if std > 0:
        skewness = np.sum(dist * ((speed_bins - mean) / std)**3)
        result['skewness'] = float(skewness)
    else:
        result['skewness'] = "NaN"
        
    # Kurtosis (excess)
    if std > 0:
        kurtosis = np.sum(dist * ((speed_bins - mean) / std)**4) - 3
        result['kurtosis'] = float(kurtosis)
    else:
        result['kurtosis'] = "NaN"
        
    # Entropy
    non_zero_dist = dist[dist > 0]
    entropy = -np.sum(non_zero_dist * np.log(non_zero_dist)) if len(non_zero_dist) > 0 else 0
    result['entropy'] = float(entropy)
    
    # Speed category percentages
    result['p_0_5'] = float(np.sum(dist[0:6]))     # 0-5 km/h
    result['p_5_15'] = float(np.sum(dist[5:16]))   # 5-15 km/h
    result['p_15_25'] = float(np.sum(dist[15:26])) # 15-25 km/h
    result['p_25_plus'] = float(np.sum(dist[25:])) # 25+ km/h
    
    return result

def wilcoxon_test_for_paired_histograms(before_means, after_means, weights, before_route_count=None, after_route_count=None, segment_count=None):
    """
    Perform Wilcoxon signed-rank test on paired data with weights.
    
    Parameters:
    -----------
    before_means : array-like
        Speed means for street segments before conversion
    after_means : array-like
        Speed means for street segments after conversion
    weights : array-like
        Weights (lengths) for each segment pair
    before_route_count : int, optional
        Total route count for before period
    after_route_count : int, optional
        Total route count for after period
    segment_count : int, optional
        Number of segments
        
    Returns:
    --------
    dict
        Dictionary containing:
        - Test statistic
        - p-value
        - Cohen's d (effect size)
        - Effective sample size
        - Percentage of segments with reduced speed
    """
    # Convert inputs to numpy arrays
    before_means = np.array(before_means)
    after_means = np.array(after_means)
    weights = np.array(weights)
    
    # Calculate differences (after - before)
    differences = after_means - before_means
    
    # Calculate effective sample size
    # If we have the new route count parameters, use them
    if before_route_count is not None and after_route_count is not None and segment_count is not None and segment_count > 0:
        before_n = before_route_count / segment_count
        after_n = after_route_count / segment_count
        effective_n = (before_n + after_n) / 2
    else:
        # Fall back to the original method if new parameters are not provided
        effective_n = calculate_effective_sample_size(weights)
    
    # Perform standard Wilcoxon signed-rank test 
    # SciPy's wilcoxon doesn't directly support weights, so we'll interpret results with caution
    w_stat, p_value = stats.wilcoxon(before_means, after_means)
    
    # Calculate Cohen's d for paired data
    cohens_d, mean_diff = calculate_cohens_d_paired(differences, weights, effective_n)
    
    # Interpret Cohen's d
    effect_interpretation = "small" if cohens_d < 0.2 else "medium" if cohens_d < 0.5 else "large"
    
    # Calculate percent of segments with reduced speed (after < before)
    pct_reduced = 100 * np.sum(after_means < before_means) / len(differences)
    
    # Calculate median difference
    median_diff = np.median(differences)
    
    # Calculate mean difference
    mean_diff = np.mean(differences)
    weighted_mean_diff = np.average(differences, weights=weights)
    
    return {
        'w_stat': w_stat,
        'p_value': p_value,
        'cohens_d': cohens_d,
        'effect_interpretation': effect_interpretation,
        'effective_n': effective_n,
        'pct_reduced': pct_reduced,
        'median_diff': median_diff,
        'mean_diff': mean_diff,
        'weighted_mean_diff': weighted_mean_diff
    }

def wkt_to_geojson(wkt_str):
    """Convert Well-Known Text (WKT) geometry to GeoJSON format"""
    try:
        # Simple WKT to GeoJSON conversion for LINESTRING
        if wkt_str.startswith('LINESTRING'):
            # Extract coordinates from LINESTRING format
            # Get the part between parentheses
            coords_str = wkt_str.split('(', 1)[1].rsplit(')', 1)[0]
            
            # Parse coordinate pairs
            coords = []
            for pair in coords_str.split(','):
                # Each pair should be "X Y" format
                parts = pair.strip().split()
                if len(parts) >= 2:
                    coords.append([float(parts[0]), float(parts[1])])
            
            return {
                'type': 'LineString',
                'coordinates': coords
            }
        elif wkt_str.startswith('MULTILINESTRING'):
            # Extract coordinates from MULTILINESTRING format
            # Get the part between double parentheses
            multi_coords_str = wkt_str.split('((', 1)[1].rsplit('))', 1)[0]
            
            # Split into individual linestrings
            linestrings = multi_coords_str.split('),(')
            
            # Parse coordinate pairs for each linestring
            multi_coords = []
            for linestring in linestrings:
                coords = []
                for pair in linestring.split(','):
                    # Each pair should be "X Y" format
                    parts = pair.strip().split()
                    if len(parts) >= 2:
                        coords.append([float(parts[0]), float(parts[1])])
                multi_coords.append(coords)
            
            return {
                'type': 'MultiLineString',
                'coordinates': multi_coords
            }
        else:
            # Return a simple point for unsupported geometry types
            log(f"Unsupported WKT geometry type: {wkt_str[:20]}...")
            return {
                'type': 'Point',
                'coordinates': [0, 0]  # Default point at origin
            }
    except Exception as e:
        log(f"Error parsing WKT string: {e} - {wkt_str[:30]}...")
        # Return a default empty geometry
        return {
            'type': 'Point',
            'coordinates': [0, 0]
        }

def create_histogram_plot(fahrradstrasse_name, paired_data, before_hist_file, after_hist_file, comparison_metrics, wilcoxon_results):
    """Create histogram plot for comparing before and after speed distributions"""
    log(f"Creating histogram plot for {fahrradstrasse_name}")
    
    # Get period information
    before_period = comparison_metrics['before_period']
    after_period = comparison_metrics['after_period']
    paired_count = comparison_metrics['paired_segments_count']
    before_route_count = comparison_metrics['before_route_count']
    after_route_count = comparison_metrics['after_route_count']
    
    # Get pre-calculated n values
    before_n = comparison_metrics['before_n']
    after_n = comparison_metrics['after_n']
    
    # Create figure
    plt.figure(figsize=PLOT_CONFIG['FIGURE_SIZE'])
    
    # Load histogram data
    before_hist_df = pd.read_csv(before_hist_file)
    after_hist_df = pd.read_csv(after_hist_file)
    
    # Get length information
    before_length_km = comparison_metrics['before_total_length'] / 1000  # Convert to m to km
    after_length_km = comparison_metrics['after_total_length'] / 1000  # Convert to m to km
    
    # Calculate means
    before_mean = np.sum(before_hist_df['speed'] * before_hist_df['frequency'])
    after_mean = np.sum(after_hist_df['speed'] * after_hist_df['frequency'])
    
    # IMPORTANT: Plot the green bars FIRST, followed by blue
    # This is the opposite order of your example but necessary to match the appearance
    plt.bar(after_hist_df['speed'], after_hist_df['frequency'] * 100, 
            alpha=0.7,  # Use 0.7 for green bars
            label=f'Nachher ({after_length_km:.1f} km, {after_route_count:,} Messwerte)', 
            color='green', 
            width=0.8)
    
    plt.bar(before_hist_df['speed'], before_hist_df['frequency'] * 100, 
            alpha=0.5,  # Use 0.5 for blue bars
            label=f'Vorher ({before_length_km:.1f} km, {before_route_count:,} Messwerte)', 
            color='blue', 
            width=0.8)
    
    # Add vertical lines for means
    plt.axvline(x=after_mean, 
                color='darkgreen', 
                linestyle='-', 
                label=f'Nachher Ø: {after_mean:.2f} km/h')
    
    plt.axvline(x=before_mean, 
                color='darkblue', 
                linestyle='-', 
                label=f'Vorher Ø: {before_mean:.2f} km/h')
    
    # Calculate difference
    diff = after_mean - before_mean
    pct_diff = 100 * diff / before_mean if before_mean != 0 else 0
    diff_text = f"Unterschied: {diff:.2f} km/h ({pct_diff:.1f}%)"
    
    # Add legend at upper left
    plt.legend(loc='upper left', fontsize=9)
    
    # Format p-value text
    p_value_text = f"p<0.001" if wilcoxon_results['p_value'] < 0.001 else f"p={wilcoxon_results['p_value']:.3f}"
    significance_text = "(Signifikant)" if wilcoxon_results['p_value'] < 0.05 else "(Nicht Signifikant)"
    
    # Add test results in a box
    plt.annotate(
        f"Wilcoxon: {p_value_text} {significance_text}\n"
        f"Cohen's d: {wilcoxon_results['cohens_d']:.3f} ({wilcoxon_results['effect_interpretation']})\n"
        f"Effective n: {wilcoxon_results['effective_n']:.1f}", 
        xy=(0.03, 0.3),
        xycoords='axes fraction',
        fontsize=10,
        bbox={'boxstyle': 'round,pad=0.5', 'fc': 'white', 'ec': 'gray', 'alpha': 0.9},
        verticalalignment='top'
    )
    
    # Add difference annotation in a box
    plt.annotate(
        diff_text, 
        xy=(0.7, 0.9),
        xycoords='axes fraction',
        fontsize=12,
        fontweight='bold',
        bbox={'boxstyle': 'round,pad=0.5', 'fc': 'white', 'ec': 'gray', 'alpha': 0.9}
    )
    
    # Add labels and styling
    plt.xlabel(PLOT_CONFIG['HIST_X_LABEL'])
    plt.ylabel(PLOT_CONFIG['HIST_Y_LABEL'])
    plt.title(PLOT_CONFIG['HIST_TITLE'])
    plt.grid(alpha=0.3)
    plt.xticks(range(0, 32, 2))
    
    # Main title and period subtitle
    plt.suptitle(
        PLOT_CONFIG['TITLE_MAIN'].format(name=fahrradstrasse_name), 
        fontsize=16, 
        fontweight='bold'
    )
    
    # Add subtitle with segment count and route counts
    n_text = f" (Vorher n={before_n:.1f}, Nachher n={after_n:.1f})"
    plt.figtext(
        0.5, 0.93,
        PLOT_CONFIG['TITLE_SUBTITLE'].format(
            count=paired_count,
            route_count_before=before_route_count,
            route_count_after=after_route_count
        ),
        ha='center',
        fontsize=12
    )
    
    # Add time period subtitle
    plt.figtext(
        0.5, 0.89,
        PLOT_CONFIG['TITLE_PERIOD'].format(before_period=before_period, after_period=after_period),
        ha='center',
        fontsize=10,
        fontstyle='italic'
    )
    
    # Apply layout adjustments
    plt.subplots_adjust(left=0.1, bottom=0.18)
    plt.tight_layout(rect=[0, 0, 1, 0.85])
    
    # Save the figure
    hist_filename = os.path.join(OUTPUT_DIR, f"{fahrradstrasse_name}_histogram.png")
    plt.savefig(
        hist_filename, 
        dpi=PLOT_CONFIG['DPI'],
        bbox_inches='tight'
    )
    plt.show()
    plt.close()
    
    log(f"Saved histogram plot to {hist_filename}")

def create_differences_plot(fahrradstrasse_name, paired_data, comparison_metrics, wilcoxon_results):
    """Create difference histogram plot showing the distribution of speed changes"""
    log(f"Creating differences plot for {fahrradstrasse_name}")
    
    # Get period information
    before_period = comparison_metrics['before_period']
    after_period = comparison_metrics['after_period']
    paired_count = comparison_metrics['paired_segments_count']
    before_route_count = comparison_metrics['before_route_count']
    after_route_count = comparison_metrics['after_route_count']
    
    # Get pre-calculated n values
    before_n = comparison_metrics['before_n']
    after_n = comparison_metrics['after_n']
    
    # Create figure
    plt.figure(figsize=PLOT_CONFIG['FIGURE_SIZE'])
    
    # Calculate differences (after - before)
    differences = np.array(paired_data['after_mean']) - np.array(paired_data['before_mean'])
    
    # Create histogram of differences
    plt.hist(differences, bins=20, color=PLOT_CONFIG['DIFF_COLOR'], alpha=0.7)
    
    # Add reference lines
    plt.axvline(x=0, 
                color=PLOT_CONFIG['NO_CHANGE_COLOR'], 
                linestyle='--', 
                label=PLOT_CONFIG['DIFF_NO_CHANGE'])
    
    plt.axvline(x=np.mean(differences), 
                color=PLOT_CONFIG['MEAN_COLOR'], 
                linestyle='-', 
                label=PLOT_CONFIG['DIFF_MEAN'].format(mean=np.mean(differences)))
    
    plt.axvline(x=np.median(differences), 
                color=PLOT_CONFIG['MEDIAN_COLOR'], 
                linestyle='-', 
                label=PLOT_CONFIG['DIFF_MEDIAN'].format(median=np.median(differences)))
    
    # Calculate weighted mean diff
    weighted_mean_diff = np.average(differences, weights=paired_data['length'])
    plt.axvline(x=weighted_mean_diff, 
                color=PLOT_CONFIG['WEIGHTED_COLOR'], 
                linestyle='-') 
                #label=PLOT_CONFIG['DIFF_WEIGHTED'].format(weighted=weighted_mean_diff))
    
    # Add labels and styling
    plt.xlabel(PLOT_CONFIG['DIFF_X_LABEL'])
    plt.ylabel(PLOT_CONFIG['DIFF_Y_LABEL'])
    plt.title(PLOT_CONFIG['DIFF_TITLE'])
    plt.legend(loc='best', fontsize=9)
    plt.grid(alpha=0.3)
    
    # Add annotation with statistical results
    plt.annotate(
        f"Wilcoxon: p={'<0.001' if wilcoxon_results['p_value'] < 0.001 else f'{wilcoxon_results['p_value']:.3f}'}\n"
        f"Segmente schneller/langsamer: {100-wilcoxon_results['pct_reduced']:.1f}% / {wilcoxon_results['pct_reduced']:.1f}%",
        xy=(0.05, 0.95),
        xycoords='axes fraction',
        fontsize=10,
        bbox=PLOT_CONFIG['TEXT_BOX_STYLE']
    )
    
    # Main title and period subtitle
    plt.suptitle(
        PLOT_CONFIG['TITLE_MAIN'].format(name=fahrradstrasse_name), 
        fontsize=16, 
        fontweight='bold'
    )
    
    # Add subtitle with segment count and route counts
    n_text = f" (Vorher n={before_n:.1f}, Nachher n={after_n:.1f})"
    plt.figtext(
        0.5, 0.93,  # Position just below main title
        PLOT_CONFIG['TITLE_SUBTITLE'].format(
            count=paired_count,
            route_count_before=before_route_count,
            route_count_after=after_route_count
        ),
        ha='center',
        fontsize=12
    )
    
    # Add time period subtitle
    plt.figtext(
        0.5, 0.89,  # Position below the count subtitle
        PLOT_CONFIG['TITLE_PERIOD'].format(before_period=before_period, after_period=after_period),
        ha='center',
        fontsize=10,
        fontstyle='italic'
    )
    
    plt.tight_layout(rect=[0, 0, 1, 0.85])  # Make room for titles
    
    # Save the figure
    diff_filename = os.path.join(OUTPUT_DIR, f"{fahrradstrasse_name}_differences.png")
    plt.savefig(
        diff_filename, 
        dpi=PLOT_CONFIG['DPI'],
        bbox_inches='tight'
    )
    plt.show()
    plt.close()
    
    log(f"Saved differences plot to {diff_filename}")

def create_scatter_plot(fahrradstrasse_name, paired_data, comparison_metrics, wilcoxon_results):
    """Create scatter plot of before vs after speeds"""
    log(f"Creating scatter plot for {fahrradstrasse_name}")
    
    # Get period information
    before_period = comparison_metrics['before_period']
    after_period = comparison_metrics['after_period']
    paired_count = comparison_metrics['paired_segments_count']
    before_route_count = comparison_metrics['before_route_count']
    after_route_count = comparison_metrics['after_route_count']
    
    # Get pre-calculated n values
    before_n = comparison_metrics['before_n']
    after_n = comparison_metrics['after_n']
    
    # Create figure
    plt.figure(figsize=PLOT_CONFIG['FIGURE_SIZE'])
    
    # Create scatter plot with segment length as point size
    min_length = min(paired_data['length'])
    max_length = max(paired_data['length'])
    normalized_lengths = [(length - min_length) / (max_length - min_length) * 100 + 10 
                          for length in paired_data['length']]
    
    plt.scatter(paired_data['before_mean'], paired_data['after_mean'], 
               alpha=PLOT_CONFIG['SCATTER_ALPHA'], 
               c=PLOT_CONFIG['SCATTER_COLOR'], 
               s=normalized_lengths)
    
    # Add identity line (y=x)
    max_speed = max(max(paired_data['before_mean']), max(paired_data['after_mean'])) + 1
    min_speed = min(min(paired_data['before_mean']), min(paired_data['after_mean'])) - 1
    plt.plot([min_speed, max_speed], [min_speed, max_speed], 
             'r--', 
             label=PLOT_CONFIG['SCATTER_EQUAL'])
    
    # Add regression line
    from scipy import stats
    slope, intercept, r_value, p_value, std_err = stats.linregress(paired_data['before_mean'], 
                                                                  paired_data['after_mean'])
    plt.plot([min_speed, max_speed], 
             [intercept + slope * min_speed, intercept + slope * max_speed], 
             'g-', 
             label=PLOT_CONFIG['SCATTER_REGRESSION'].format(
                 slope=slope, 
                 intercept=intercept, 
                 r_squared=r_value**2
             ))
    
    # Add labels and styling
    plt.xlabel(PLOT_CONFIG['SCATTER_X_LABEL'])
    plt.ylabel(PLOT_CONFIG['SCATTER_Y_LABEL'])
    plt.title(PLOT_CONFIG['SCATTER_TITLE'])
    plt.legend(loc='best', fontsize=9)
    plt.grid(alpha=0.3)
    
    # Add annotation with count information
    plt.annotate(
        f"Punkte: {paired_count} Segmente\nPunktgröße entspricht Segmentlänge",
        xy=(0.05, 0.95),
        xycoords='axes fraction',
        fontsize=10,
        bbox=PLOT_CONFIG['TEXT_BOX_STYLE']
    )
    
    # Main title and period subtitle
    plt.suptitle(
        PLOT_CONFIG['TITLE_MAIN'].format(name=fahrradstrasse_name), 
        fontsize=16, 
        fontweight='bold'
    )
    
    # Add subtitle with segment count and route counts
    n_text = f" (Vorher n={before_n:.1f}, Nachher n={after_n:.1f})"
    plt.figtext(
        0.5, 0.93,  # Position just below main title
        PLOT_CONFIG['TITLE_SUBTITLE'].format(
            count=paired_count,
            route_count_before=before_route_count,
            route_count_after=after_route_count
        ),
        ha='center',
        fontsize=12
    )
    
    # Add time period subtitle
    plt.figtext(
        0.5, 0.89,  # Position below the count subtitle
        PLOT_CONFIG['TITLE_PERIOD'].format(before_period=before_period, after_period=after_period),
        ha='center',
        fontsize=10,
        fontstyle='italic'
    )
    
    plt.tight_layout(rect=[0, 0, 1, 0.85])  # Make room for titles
    
    # Save the figure
    scatter_filename = os.path.join(OUTPUT_DIR, f"{fahrradstrasse_name}_scatter.png")
    plt.savefig(
        scatter_filename, 
        dpi=PLOT_CONFIG['DPI'],
        bbox_inches='tight'
    )
    plt.show()
    plt.close()
    
    log(f"Saved scatter plot to {scatter_filename}")

def create_boxplot(fahrradstrasse_name, paired_data, comparison_metrics, wilcoxon_results):
    """Create boxplot comparing before and after distributions"""
    log(f"Creating boxplot for {fahrradstrasse_name}")
    
    # Get period information
    before_period = comparison_metrics['before_period']
    after_period = comparison_metrics['after_period']
    paired_count = comparison_metrics['paired_segments_count']
    before_route_count = comparison_metrics['before_route_count']
    after_route_count = comparison_metrics['after_route_count']
    
    # Get pre-calculated n values
    before_n = comparison_metrics['before_n']
    after_n = comparison_metrics['after_n']
    
    # Create figure
    plt.figure(figsize=PLOT_CONFIG['FIGURE_SIZE'])
    
    # Create box plots - using paired data means
    boxplot = plt.boxplot([paired_data['before_mean'], paired_data['after_mean']], 
                         labels=[PLOT_CONFIG['BOX_BEFORE_LABEL'] + f" (n={before_n:.1f})", 
                                 PLOT_CONFIG['BOX_AFTER_LABEL'] + f" (n={after_n:.1f})"], 
                         patch_artist=True, 
                         showfliers=False)
    
    # Set colors for boxes
    boxplot['boxes'][0].set_facecolor(PLOT_CONFIG['BOX_BEFORE_COLOR'])
    boxplot['boxes'][0].set_alpha(PLOT_CONFIG['BOX_ALPHA'])
    boxplot['boxes'][1].set_facecolor(PLOT_CONFIG['BOX_AFTER_COLOR'])
    boxplot['boxes'][1].set_alpha(PLOT_CONFIG['BOX_ALPHA'])
    
    # Add individual points for better visualization
    y_offset = 0.05  # Small offset to separate points visually
    for i, values in enumerate([paired_data['before_mean'], paired_data['after_mean']]):
        # Add jittered points
        for value in values:
            plt.plot(i+1 + (np.random.random()-0.5)*0.2, value, 'o', 
                    color=PLOT_CONFIG['BEFORE_COLOR_DARK'] if i==0 else PLOT_CONFIG['AFTER_COLOR_DARK'], 
                    alpha=0.4, 
                    markersize=4)
    
    # Add lines connecting paired points
    for i in range(len(paired_data['before_mean'])):
        plt.plot([1, 2], [paired_data['before_mean'][i], paired_data['after_mean'][i]], 
                'gray', alpha=0.1, linewidth=0.5)
    
    # Calculate weight-based means
    before_weighted_mean = np.average(paired_data['before_mean'], weights=paired_data['before_weight'])
    after_weighted_mean = np.average(paired_data['after_mean'], weights=paired_data['after_weight'])
    
    # Add weighted means as separate markers
    plt.plot(1, before_weighted_mean, '*', 
             color=PLOT_CONFIG['BEFORE_COLOR_DARK'], 
             markersize=15, 
             label=PLOT_CONFIG['BOX_WEIGHTED_MEAN'].format(mean=before_weighted_mean))
    
    plt.plot(2, after_weighted_mean, '*', 
             color=PLOT_CONFIG['AFTER_COLOR_DARK'], 
             markersize=15, 
             label=PLOT_CONFIG['BOX_WEIGHTED_MEAN'].format(mean=after_weighted_mean))
    
    # Add labels and styling
    plt.ylabel(PLOT_CONFIG['BOX_Y_LABEL'])
    plt.title(PLOT_CONFIG['BOX_TITLE'])
    plt.grid(axis='y', alpha=0.3)
    plt.legend(loc='upper right', fontsize=9)

    # Set y-axis limits to 8-24 km/h
    plt.ylim(8, 24)
    
    # Add effective sample size information
    plt.annotate(
        f"Segmente: {paired_count}", 
        xy=(0.05, 0.95), 
        xycoords='axes fraction',
        fontsize=10,
        bbox=PLOT_CONFIG['TEXT_BOX_STYLE']
    )
    
    # Add annotation with statistical test results
    plt.annotate(
        f"Wilcoxon p-value: {'<0.001' if wilcoxon_results['p_value'] < 0.001 else f'{wilcoxon_results['p_value']:.3f}'} "
        f"({'signifikant' if wilcoxon_results['p_value'] < 0.05 else 'nicht signifikant'})\n"
        f"Unterschied Mittelwert: {after_weighted_mean - before_weighted_mean:.2f} km/h", 
        xy=(0.05, 0.2), 
        xycoords='axes fraction',
        fontsize=10,
        bbox=PLOT_CONFIG['TEXT_BOX_STYLE']
    )
    
    # Main title and period subtitle
    plt.suptitle(
        PLOT_CONFIG['TITLE_MAIN'].format(name=fahrradstrasse_name), 
        fontsize=16, 
        fontweight='bold'
    )
    
    # Add subtitle with segment count and route counts
    n_text = f" (Vorher n={before_n:.1f}, Nachher n={after_n:.1f})"
    plt.figtext(
        0.5, 0.93,  # Position just below main title
        PLOT_CONFIG['TITLE_SUBTITLE'].format(
            count=paired_count,
            route_count_before=before_route_count,
            route_count_after=after_route_count
        ),
        ha='center',
        fontsize=12
    )
    
    # Add time period subtitle
    plt.figtext(
        0.5, 0.89,  # Position below the count subtitle
        PLOT_CONFIG['TITLE_PERIOD'].format(before_period=before_period, after_period=after_period),
        ha='center',
        fontsize=10,
        fontstyle='italic'
    )
    
    plt.tight_layout(rect=[0, 0, 1, 0.85])  # Make room for titles
    
    # Save the figure
    boxplot_filename = os.path.join(OUTPUT_DIR, f"{fahrradstrasse_name}_boxplot.png")
    plt.savefig(
        boxplot_filename, 
        dpi=PLOT_CONFIG['DPI'],
        bbox_inches='tight'
    )
    plt.show()
    plt.close()
    
    log(f"Saved boxplot to {boxplot_filename}")

def process_fahrradstrasse_before_after_paired(parquet_file):
    """Process before-after periods for each Fahrradstraße with paired analysis"""
    log("Processing before-after periods for each Fahrradstraße with paired analysis")
    
    # Initialize storage for results
    results = []
    segment_data_collection = []
    
    # Get all available columns from the parquet file
    all_columns = [field.name for field in parquet_file.schema]
    
    # Check if geometry column exists
    has_geometry = COLUMN_GEOMETRY in all_columns
    if not has_geometry:
        log(f"Warning: Geometry column '{COLUMN_GEOMETRY}' not found in parquet file.")
    
    # Process each Fahrradstraße
    for fahrradstrasse_name, periods in FAHRRADSTRASSE_PERIODS.items():
        log(f"Processing {fahrradstrasse_name}")
        
        # Get the column that identifies this Fahrradstraße
        fahrradstrasse_col = FAHRRADSTRASSE_COLUMNS[fahrradstrasse_name]
        
        # Format period text for plot titles
        before_period_text = format_period_text(periods['before'])
        after_period_text = format_period_text(periods['after'])
        
        # Initialize storage for paired segment data
        paired_data = {
            'segment_id': [],  # Unique segment identifiers
            'before_hist': [], # Aggregated before histogram for each segment
            'after_hist': [],  # Aggregated after histogram for each segment
            'before_mean': [], # Mean speed from before histogram
            'after_mean': [],  # Mean speed from after histogram
            'length': [],      # Segment length (same for before and after)
            'before_length_weighted_hist': [], # Length-weighted before histogram
            'after_length_weighted_hist': [],  # Length-weighted after histogram
            'geometry': [],    # WKT geometry if available
            'before_route_count': [], # Total route count for before period
            'after_route_count': [],  # Total route count for after period
            'before_weight': [], # Weight for before period (based on selected weighting)
            'after_weight': []   # Weight for after period (based on selected weighting)
        }
        
        # Initialize storage for aggregated histograms
        agg_data = {
            'before': {'histograms': [], 'weights': [], 'total_length': 0, 'count': 0, 'total_route_count': 0},
            'after': {'histograms': [], 'weights': [], 'total_length': 0, 'count': 0, 'total_route_count': 0}
        }
        
        # Find period histogram columns and route count columns
        period_hist_cols = {'before': [], 'after': []}
        period_route_count_cols = {'before': [], 'after': []}
        
        for period_name, months in periods.items():
            log(f"  Processing {period_name} period: {', '.join(months)}")
            
            # Find histogram columns and route count columns for these months
            for month in months:
                # Speed histogram column
                exact_col = f"{month}_speeds"
                if exact_col in all_columns:
                    period_hist_cols[period_name].append(exact_col)
                
                # Route count column
                route_count_col = get_route_count_column(month)
                if route_count_col in all_columns:
                    period_route_count_cols[period_name].append(route_count_col)
            
            if not period_hist_cols[period_name]:
                log(f"Warning: No exact histogram columns found for {period_name} period {months}")
                continue
            
            log(f"    Found {len(period_hist_cols[period_name])} histogram columns and {len(period_route_count_cols[period_name])} route count columns")
        
        # Read all necessary columns
        columns_to_read = [
            fahrradstrasse_col, 
            COLUMN_LENGTH, 
            'id'  # Using id as the segment identifier
        ] + period_hist_cols['before'] + period_hist_cols['after'] + period_route_count_cols['before'] + period_route_count_cols['after']
        
        # Add geometry column if available
        if has_geometry:
            columns_to_read.append(COLUMN_GEOMETRY)
        # Process each row group
        num_row_groups = parquet_file.metadata.num_row_groups
        for rg in tqdm(range(num_row_groups), desc=f"Processing {fahrradstrasse_name}"):
            try:
                # Read the columns for this row group
                table = parquet_file.read_row_group(rg, columns=columns_to_read)
                df_chunk = table.to_pandas()
                
                # Apply filters in pandas
                df_chunk = df_chunk[
                    (df_chunk[fahrradstrasse_col] == 1) & 
                    (df_chunk[COLUMN_LENGTH] > 0)
                ]
                
                if df_chunk.empty:
                    continue
                
                # For each row, combine all histograms from before and after periods
                for _, row in df_chunk.iterrows():
                    segment_id = row['id']
                    segment_length = row[COLUMN_LENGTH]
                    segment_geometry = row.get(COLUMN_GEOMETRY, None) if has_geometry else None
                    
                    # Process before histograms
                    before_hists = []
                    for col in period_hist_cols['before']:
                        hist = parse_histogram(row[col])
                        if len(hist) > 0 and np.sum(hist) > 0:
                            before_hists.append(hist)
                    
                    # Process after histograms
                    after_hists = []
                    for col in period_hist_cols['after']:
                        hist = parse_histogram(row[col])
                        if len(hist) > 0 and np.sum(hist) > 0:
                            after_hists.append(hist)
                    
                    # Calculate route counts for both periods
                    before_route_count = 0
                    for col in period_route_count_cols['before']:
                        if col in row and not pd.isna(row[col]):
                            before_route_count += row[col]
                            
                    after_route_count = 0
                    for col in period_route_count_cols['after']:
                        if col in row and not pd.isna(row[col]):
                            after_route_count += row[col]
                    
                    # Calculate weights based on selected weight type
                    before_weight = get_weight_factor(row, periods['before'], WEIGHT_TYPE)
                    after_weight = get_weight_factor(row, periods['after'], WEIGHT_TYPE)
                    
                    # Only store data if we have both before and after histograms
                    if before_hists and after_hists:
                        # Average histograms for each period
                        before_avg_hist = np.mean(before_hists, axis=0)
                        after_avg_hist = np.mean(after_hists, axis=0)
                        
                        # Calculate mean speeds from histograms
                        before_mean = calculate_weighted_mean(before_avg_hist)
                        after_mean = calculate_weighted_mean(after_avg_hist)
                        
                        # Only include if both means could be calculated
                        if not np.isnan(before_mean) and not np.isnan(after_mean):
                            # Store the segment data
                            paired_data['segment_id'].append(segment_id)
                            paired_data['before_hist'].append(before_avg_hist)
                            paired_data['after_hist'].append(after_avg_hist)
                            paired_data['before_mean'].append(before_mean)
                            paired_data['after_mean'].append(after_mean)
                            paired_data['length'].append(segment_length)
                            paired_data['geometry'].append(segment_geometry)
                            paired_data['before_route_count'].append(before_route_count)
                            paired_data['after_route_count'].append(after_route_count)
                            paired_data['before_weight'].append(before_weight)
                            paired_data['after_weight'].append(after_weight)
                            
                            # Store weighted histograms based on selected weight type
                            paired_data['before_length_weighted_hist'].append(before_avg_hist * before_weight)
                            paired_data['after_length_weighted_hist'].append(after_avg_hist * after_weight)
                            
                            # Add to aggregated data
                            agg_data['before']['histograms'].append(before_avg_hist)
                            agg_data['before']['weights'].append(before_weight)
                            agg_data['before']['total_length'] += segment_length
                            agg_data['before']['count'] += 1
                            agg_data['before']['total_route_count'] += before_route_count
                            
                            agg_data['after']['histograms'].append(after_avg_hist)
                            agg_data['after']['weights'].append(after_weight)
                            agg_data['after']['total_length'] += segment_length
                            agg_data['after']['count'] += 1
                            agg_data['after']['total_route_count'] += after_route_count
                
                # Clean up
                del df_chunk, table
                if rg % 5 == 0:
                    force_gc()
                    
            except Exception as e:
                log(f"Error processing row group {rg}: {e}")
        # Process paired data if we have enough segments
        if len(paired_data['segment_id']) > 5:  # Minimum threshold for analysis
            log(f"Performing paired analysis for {fahrradstrasse_name} with {len(paired_data['segment_id'])} paired segments")
            
            # Calculate metrics from aggregated histograms
            speed_bins = np.arange(32)  # Speed bins from 0 to 31 km/h
            
            # Process before period for aggregated histogram metrics
            before_weighted_hists = np.array(paired_data['before_length_weighted_hist'])
            before_combined_hist = np.sum(before_weighted_hists, axis=0)
            before_dist = before_combined_hist / np.sum(before_combined_hist) if np.sum(before_combined_hist) > 0 else np.zeros_like(before_combined_hist)
            
            # Process after period for aggregated histogram metrics
            after_weighted_hists = np.array(paired_data['after_length_weighted_hist'])
            after_combined_hist = np.sum(after_weighted_hists, axis=0)
            after_dist = after_combined_hist / np.sum(after_combined_hist) if np.sum(after_combined_hist) > 0 else np.zeros_like(after_combined_hist)
            
            # Calculate metrics without bootstrap CI
            before_metrics = calculate_metrics(before_dist, speed_bins)
            after_metrics = calculate_metrics(after_dist, speed_bins)
            
            # Save raw histogram data for visualization
            before_hist_file = os.path.join(OUTPUT_DIR, f"{fahrradstrasse_name}_before_histogram.csv")
            after_hist_file = os.path.join(OUTPUT_DIR, f"{fahrradstrasse_name}_after_histogram.csv")
            
            pd.DataFrame({'speed': speed_bins, 'frequency': before_dist}).to_csv(before_hist_file, index=False)
            pd.DataFrame({'speed': speed_bins, 'frequency': after_dist}).to_csv(after_hist_file, index=False)
            
            # Calculate total route counts
            total_before_route_count = sum(paired_data['before_route_count'])
            total_after_route_count = sum(paired_data['after_route_count'])
            segment_count = len(paired_data['segment_id'])
            
            # Calculate n values once
            before_n = total_before_route_count / segment_count if segment_count > 0 else 0
            after_n = total_after_route_count / segment_count if segment_count > 0 else 0
            
            # Perform Wilcoxon signed-rank test on paired segment means
            wilcoxon_results = wilcoxon_test_for_paired_histograms(
                paired_data['before_mean'],
                paired_data['after_mean'],
                paired_data['before_weight'] if WEIGHT_TYPE != 'segment_count' else paired_data['length'],
                total_before_route_count,  # Pass this parameter
                total_after_route_count,   # Pass this parameter
                segment_count              # Pass this parameter
            )
            
            # Create comparison metrics
            comparison_metrics = {
                'fahrradstrasse': fahrradstrasse_name,
                'paired_segments_count': segment_count,
                'before_total_length': agg_data['before']['total_length'],
                'after_total_length': agg_data['after']['total_length'],
                'before_route_count': total_before_route_count,
                'after_route_count': total_after_route_count,
                'before_n': before_n,  # Add new field here
                'after_n': after_n,    # Add new field here 
                'effective_n': wilcoxon_results['effective_n'],
                'before_period': before_period_text,
                'after_period': after_period_text,
                'weight_type': WEIGHT_TYPE
            }
            
            # Add Wilcoxon test results
            for key, value in wilcoxon_results.items():
                comparison_metrics[f'wilcoxon_{key}'] = value
            
            # Add all metrics from before period with 'before_' prefix
            for metric, value in before_metrics.items():
                comparison_metrics[f'before_{metric}'] = value
            
            # Add all metrics from after period with 'after_' prefix
            for metric, value in after_metrics.items():
                comparison_metrics[f'after_{metric}'] = value
            
            # Calculate differences for key metrics
            for metric in ['mean', 'std', 'cv', 'entropy', 'p_0_5', 'p_5_15', 'p_15_25', 'p_25_plus']:
                before_value = before_metrics[metric]
                after_value = after_metrics[metric]
                
                # Skip if either value is "NaN"
                if before_value == "NaN" or after_value == "NaN":
                    comparison_metrics[f'diff_{metric}'] = "NaN"
                    comparison_metrics[f'pct_diff_{metric}'] = "NaN"
                else:
                    # Calculate absolute difference
                    abs_diff = after_value - before_value
                    comparison_metrics[f'diff_{metric}'] = abs_diff
                    
                    # Calculate percentage difference
                    if before_value != 0:
                        pct_diff = 100 * abs_diff / before_value
                        comparison_metrics[f'pct_diff_{metric}'] = pct_diff
                    else:
                        comparison_metrics[f'pct_diff_{metric}'] = "NaN"
            
            results.append(comparison_metrics)
            
            # Create paired before-after plots (separate plots)
            create_histogram_plot(
                fahrradstrasse_name, 
                paired_data, 
                before_hist_file, 
                after_hist_file, 
                comparison_metrics,
                wilcoxon_results
            )
            
            create_differences_plot(
                fahrradstrasse_name,
                paired_data,
                comparison_metrics,
                wilcoxon_results
            )
            
            create_scatter_plot(
                fahrradstrasse_name,
                paired_data,
                comparison_metrics,
                wilcoxon_results
            )
            
            create_boxplot(
                fahrradstrasse_name,
                paired_data,
                comparison_metrics,
                wilcoxon_results
            )
            # Collect segment data for combined CSV and GeoJSON
            for i, segment_id in enumerate(paired_data['segment_id']):
                segment_data = {
                    'segment_id': segment_id,
                    'fahrradstrasse': fahrradstrasse_name,
                    'length': paired_data['length'][i],
                    'before_mean': paired_data['before_mean'][i],
                    'after_mean': paired_data['after_mean'][i],
                    'before_route_count': paired_data['before_route_count'][i],
                    'after_route_count': paired_data['after_route_count'][i],
                    'diff': paired_data['after_mean'][i] - paired_data['before_mean'][i],
                    'pct_diff': 100 * (paired_data['after_mean'][i] - paired_data['before_mean'][i]) / paired_data['before_mean'][i] if paired_data['before_mean'][i] > 0 else np.nan,
                }
                
                if has_geometry and paired_data['geometry'][i] is not None:
                    segment_data['geometry_wkt'] = paired_data['geometry'][i]
                
                segment_data_collection.append(segment_data)
        else:
            log(f"Insufficient paired data for {fahrradstrasse_name}, only {len(paired_data['segment_id'])} paired segments found")
    
    # Create combined CSV and GeoJSON outputs
    if segment_data_collection:
        # Create CSV with all segment data
        segments_df = pd.DataFrame(segment_data_collection)
        segments_csv_path = os.path.join(OUTPUT_DIR, "all_fahrradstrasse_segments.csv")
        segments_df.to_csv(segments_csv_path, index=False)
        log(f"Saved all segment data to {segments_csv_path}")
        
        # Create GeoJSON if geometry data is available
        if has_geometry and 'geometry_wkt' in segments_df.columns:
            # Create GeoJSON features
            features = []
            for _, row in segments_df.iterrows():
                if pd.notna(row.get('geometry_wkt')):
                    try:
                        # Create GeoJSON feature with properties
                        geometry = wkt_to_geojson(row['geometry_wkt'])
                        
                        feature = {
                            'type': 'Feature',
                            'geometry': geometry,
                            'properties': {
                                'segment_id': str(row['segment_id']),  # Convert to string to avoid JSON issues
                                'fahrradstrasse': row['fahrradstrasse'],
                                'length': float(row['length']),
                                'before_mean': float(row['before_mean']),
                                'after_mean': float(row['after_mean']),
                                'before_route_count': int(row['before_route_count']),
                                'after_route_count': int(row['after_route_count']),
                                'diff': float(row['diff']),
                                'pct_diff': float(row['pct_diff']) if not np.isnan(row['pct_diff']) else None
                            }
                        }
                        features.append(feature)
                    except Exception as e:
                        log(f"Error creating GeoJSON feature for segment {row['segment_id']}: {e}")
                        continue  # Skip this feature, but continue with others
            
            # Create the GeoJSON object
            geojson = {
                'type': 'FeatureCollection',
                'features': features
            }
            
            # Save GeoJSON file
            geojson_path = os.path.join(OUTPUT_DIR, "all_fahrradstrasse_segments.geojson")
            with open(geojson_path, 'w') as f:
                json.dump(geojson, f)
            log(f"Saved GeoJSON data to {geojson_path}")
    
    # Compile all results into a DataFrame
    if results:
        results_df = pd.DataFrame(results)
        results_df.to_csv(os.path.join(OUTPUT_DIR, "fahrradstrasse_wilcoxon_results.csv"), index=False)
        log(f"Saved results to {os.path.join(OUTPUT_DIR, 'fahrradstrasse_wilcoxon_results.csv')}")
        return results_df, segments_df
    else:
        log("No results to save")
        return None, None

def create_central_comparison_table(results_df, segments_df):
    """Create a central comparison table summarizing all Fahrradstraße results"""
    log("Creating central comparison table")
    
    # Create a summary table with key metrics
    summary_table = []
    
    for _, row in results_df.iterrows():
        fahrradstrasse = row['fahrradstrasse']
        
        # Count segments that got faster/slower
        segments_this_fs = segments_df[segments_df['fahrradstrasse'] == fahrradstrasse]
        num_faster = sum(segments_this_fs['diff'] > 0)
        num_slower = sum(segments_this_fs['diff'] < 0)
        num_same = sum(segments_this_fs['diff'] == 0)
        before_n = row['before_n']
        after_n = row['after_n']
        
        # Create summary entry
        summary = {
            'Fahrradstrasse': fahrradstrasse,
            'Before_period': row['before_period'],
            'After_period': row['after_period'],
            'Segment_count': row['paired_segments_count'],
            'Total_length_km': row['before_total_length'] / 1000,
            'Before_route_count': row['before_route_count'],
            'After_route_count': row['after_route_count'],
            'Effective_sample_size': row['effective_n'],
            'Before_n': before_n,
            'After_n': after_n,
            'Before_mean_speed': row['before_mean'],
            'After_mean_speed': row['after_mean'],
            'Speed_difference': row['diff_mean'],
            'Speed_difference_pct': row['pct_diff_mean'],
            'Wilcoxon_p_value': row['wilcoxon_p_value'],
            'Significant': row['wilcoxon_p_value'] < 0.05,
            'Cohens_d': row['wilcoxon_cohens_d'],
            'Effect_size': row['wilcoxon_effect_interpretation'],
            'Segments_faster': num_faster,
            'Segments_slower': num_slower,
            'Segments_unchanged': num_same,
            'Pct_segments_slower': row['wilcoxon_pct_reduced']
        }
        
        summary_table.append(summary)
    
    # Convert to DataFrame and save
    summary_df = pd.DataFrame(summary_table)
    summary_csv_path = os.path.join(OUTPUT_DIR, "fahrradstrasse_central_comparison.csv")
    summary_df.to_csv(summary_csv_path, index=False)
    log(f"Saved central comparison table to {summary_csv_path}")
    
    return summary_df

def run_fahrradstrasse_before_after_paired_analysis():
    """Run the paired before-after analysis for Fahrradstraßen"""
    try:
        log("Starting Fahrradstraße Paired Before-After Analysis with Wilcoxon Signed-Rank Test")
        log(f"Using weight type: {WEIGHT_TYPE}")
        
        # Open the parquet file
        parquet_file = pq.ParquetFile(PARQUET_FILE)
        
        # Process before-after periods with paired analysis
        results_df, segments_df = process_fahrradstrasse_before_after_paired(parquet_file)
        
        # Create a central comparison table with all results
        if results_df is not None:
            # Create a central summary table
            summary_table = create_central_comparison_table(results_df, segments_df)
            
            log("Paired Before-After analysis complete! Results saved to: " + OUTPUT_DIR)
            return results_df, segments_df
        else:
            log("No results to analyze")
            return None, None
    
    except Exception as e:
        log(f"Error in Paired Before-After analysis: {e}")
        import traceback
        traceback.print_exc()
        return None, None

# Run the analysis
if __name__ == "__main__":
    run_fahrradstrasse_before_after_paired_analysis()
```
